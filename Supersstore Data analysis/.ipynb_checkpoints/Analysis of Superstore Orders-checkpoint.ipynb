#Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#Read data

```python
dataset = pd.read_excel("Superstore_USA.xlsx")
```

#Explore the dataset

```python
dataset.head()
```

```python
dataset.shape
```

#Find the null values present in dataset

```python
dataset.isnull().sum()
```

#fill null values in dataset with mean value

```python
dataset.fillna({'Product Base Margin': dataset['Product Base Margin'].mean()}, inplace=True)
```

```python
dataset.isnull().sum()
```

#explore Order Priority frequncy

```python
dataset['Order Priority'].value_counts()
```

```python
dataset['Order Priority'].unique()
```

```python
dataset['Order Priority'] = dataset['Order Priority'].replace("Critical ","Critical")
```

```python
dataset['Order Priority'].value_counts()
```

Plot the graph for count of Order Priority

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Order Priority", data=dataset)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")
plt.show()
```

#Analyzing Shipping Mode order counts

```python
dataset['Ship Mode'].value_counts()
```

```python
x = dataset['Ship Mode'].value_counts().index
y = dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y, labels=x,startangle= 60, autopct="%0.2f%%")
plt.legend(loc=2)
plt.savefig("Percentage of orders shipping by mode.jpg")
plt.show()
```

#Analyzing Shipping Mode order counts by the product category

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Ship Mode", data=dataset,hue="Product Category")
plt.savefig("Frequency of shipping mode by category.jpg")
plt.show()
```

#Analyzing which segment is placing more orders

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Customer Segment", data=dataset)
plt.savefig("Count of orders by Customer Segment.jpg")
plt.show()
```

#categories with frequency of customers

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Product Category", data=dataset)
plt.savefig("Count of orders by product category.jpg")
plt.show()
```

#customers count in segment of office supplies sub category wise

```python
plt.figure(figsize=(13,7))
sns.countplot(x="Product Category", data=dataset[dataset['Product Category']=="Office Supplies"],hue="Product Sub-Category")
plt.savefig("frequency of office supplies products by sub-category.jpg")
plt.show()
```

#customers count in segment of technology sub category wise

```python
plt.figure(figsize=(13,7))
sns.countplot(x="Product Category", data=dataset[dataset['Product Category']=="Technology"],hue="Product Sub-Category")
plt.savefig("frequency of technology products by sub-category.jpg")
plt.show()
```

#customers count in segment of furniture sub category wise

```python
plt.figure(figsize=(13,7))
sns.countplot(x="Product Category", data=dataset[dataset['Product Category']=="Furniture"],hue="Product Sub-Category")
plt.savefig("frequency of furniture products by sub-category.jpg")
plt.show()
```

```python
dataset.info()
```

#Extracting only Year from the Date column and adding to new column

```python
dataset["Order Year"] = dataset['Order Date'].dt.year
```

```python
dataset.info()
```

```python
dataset['Order Year'].value_counts()
```

#Now analyzing order frequency by year

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Order Year", data=dataset)
plt.savefig("Order count by year.jpg")
plt.show()
```

#plotting sum of profits made by the Product category

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Product Category", y="Profit",data=dataset,estimator="sum")
plt.savefig("sum of profit by product category.jpg")
plt.show()
```
